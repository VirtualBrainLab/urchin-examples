```python
import oursin as urchin
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import tifffile

# Load TIFF file
volume = tifffile.imread('./180min_cfos_stack.tiff')

# Check the shape of the loaded volume
print(volume.shape)
```

```python
volume[volume==0] = np.nan
```

```python
# rotate volume to ap/ml/dv
volume = np.transpose(volume, (0,2,1))
plt.imshow(volume[250,:,:]) # this should be a coronal slice
```

```python

plt.imshow(volume[:,150,:]) # this should be a sagittal slice
```

```python
plt.imshow(volume[150,:,:])
```

```python
urchin.setup()
```

```python
urchin.camera.main.set_mode('perspective')
```

```python
compressed, quantiles = urchin.volumes.compress_volume(volume)
```

```python
colormap = urchin.volumes.colormap(datapoints=quantiles)
```

```python
dimensions = [dim/1000*25 for dim in compressed.shape]
resolution = [25, 25, 25]
```

```python
atlas = urchin.ontology.CustomAtlas("MRI", dimensions, resolution)
urchin.volumes.Volume(compressed, colormap)
```
